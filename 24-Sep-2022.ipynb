Matplotlib Matplotlib is a 2-D plotting library that helps in Visualizing figures. Matplolib emulates Matlab like graphs and visualizations.
 
Matplotlib has a module called pyplot which aids in plotting figure. the jupyter notebook is used for running the plots. We import matlpotlib.pyplot as plt for making it call the package module.

plt.plot()for ploting line chart similarly in place of plot other functions are used for plotting. all plotting functions require data and it is provided in the function through parameters. plt.xlable, plt.ylable for labling x and y-axis respectively.

```python
#from matplolib import pyplot as plt

import matplotlib.pyplot as plt

import numpy as np
```

```python
a=np.array([1,2,3,4,5,6,7])
b=np.array([1,2,3,4,5,6,7])
```

```python
plt.title("Linear Line")
plt.xlabel('a')
plt.ylabel('b')
plt.plot(a,b)
plt.show()
```

```python
g=np.array([3,6,44,12,67,45,55])
h=np.array([22,4,5,6,78,22,12])
plt.xlabel('g')
plt.ylabel('h')
plt.plot(g,h)
plt.show()
```

```python
plt.plot(g,h,'--g')
plt.show()
```

```python
x = [5,8,10,2,6,12]
y = [12,16,6,10,14,6]
plt.bar(x,y)
plt.title('Bar graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
plt.bar(x,y, color = 'g',align = 'center')
plt.title('Bar graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

Histogram

A histogram takes in series of data and divides the data into a number of bins. it then plots the frequency data points in each bin(i.e the interval of points). it
is useful in understanding the count of data ranges. Whene to use histogram whene we need the count of the variable in plot.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plt.hist?
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins = [0,20,40,60,80,100])
plt.title("histogram")
plt.show()
```
